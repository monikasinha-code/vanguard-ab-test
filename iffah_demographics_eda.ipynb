```python
import pandas as pd

df = pd.read_csv("vanguard_master_cleaned.csv")

clients = df.drop_duplicates(subset="client_id").copy()
```

```python
# Check The Unique Client Dataset

clients.shape
```

```python
# Average age
round(clients["clnt_age"].mean(), 2)
```

```python
# Average Tenure

round(clients["clnt_tenure_yr"].mean(), 2)
```

```python
# Gender Split

clients["gendr"].value_counts()
round(clients["gendr"].value_counts(normalize=True) * 100, 2)
```

```python
# Are they new or long-standing investors?

def tenure_group(years):
    if years < 5:
        return "New client"
    elif years <= 10:
        return "Established client"
    else:
        return "Long-standing client"

clients["tenure_group"] = clients["clnt_tenure_yr"].apply(tenure_group)
```

```python
# Group counts and percentage

clients["tenure_group"].value_counts()
```

```python
# Percentage of the group

round(clients["tenure_group"].value_counts(normalize=True) * 100, 2)
```

```python
# Check high balance clients

clients["is_high_balance"].value_counts()
```

```python
# Percentage of high balance clients
round(clients["is_high_balance"].value_counts(normalize=True) * 100, 2)
```

```python
# Check multi-account clients

clients["is_multi_account"].value_counts()
```

```python
# Check percentages of multi-account clients

round(clients["is_multi_account"].value_counts(normalize=True) * 100, 2)
```

## Phase 1 Findings: Demographic Analysis

After removing duplicate client records, the dataset contains **50,488 unique clients**. This client-level dataset was used for demographic analysis to avoid counting the same client multiple times across different process steps.

The average client age is **47.32 years**, which suggests that the primary users of the Vanguard digital process are middle-aged investors. The gender distribution is fairly balanced: **34.23% unknown/unspecified**, **33.57% male**, and **32.20% female**.

The average client tenure is **12.03 years**, showing that most users are not new investors. Based on tenure groups, **51.07%** of clients are long-standing clients, **41.45%** are established clients, and only **7.48%** are new clients. This suggests that the primary clients using the process are mostly long-standing or established Vanguard investors.

For client segments, **22.98%** of unique clients are classified as high-balance clients, while **21.32%** are multi-account clients. These groups are important for the next phases because they may behave differently during the online process and may need closer analysis for drop-off points or personalized support.

Overall, the primary clients using this process appear to be middle-aged, relatively long-tenured investors, with a balanced gender distribution and a meaningful segment of high-balance and multi-account clients.
